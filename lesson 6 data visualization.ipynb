```python
import numpy as np
import pandas as pd
import os 
import matplotlib.pyplot as plt
```

```python
cars_data=pd.read_csv('ToyotaCorolla.csv',na_values=['??','???'])
```

```python
cars_data2=cars_data.select_dtypes(exclude=[object])
```

```python
cars_data2.info()
```

#drop missing values


```python
cars_data2=cars_data.copy(deep=True)
```

```python
cars_data2.dropna(axis=0,inplace=True)
```

```python
cars_data2.isnull().sum()
```

#steps of scatter plot


```python
plt.scatter(cars_data2['Age'],cars_data2['Price'],c='blue')
plt.title('Scatter plot for Price vs Age of the cars')
plt.xlabel("Age(months)")
plt.ylabel("Price(Dollar)")
```

```python
plt.show()
```

#Histogram

```python
plt.hist(cars_data2['KM'],color='Green',edgecolor='white',bins=5)
#it means most of the cars have been travelled between 50k to 1l 
plt.title('Histogram of Kilometer')
plt.xlabel('Kilometer')
plt.ylabel('Frequency')
```

```python
plt.hist(cars_data2['Price'],color='Green',edgecolor='white',bins=30)
#it means most of the cars have been travelled between 50k to 1l 
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
```
